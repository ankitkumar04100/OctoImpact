# OctoImpact AI Sustainability Insights Notebook
This notebook demonstrates the AI-driven sustainability insights generated for OctoImpact users.

## 1. Import Libraries

```python
import random
import pandas as pd
import numpy as np

# For visualization
import matplotlib.pyplot as plt
%matplotlib inline
```

## 2. Sample Sustainability Data

```python
data = {
    'User': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Energy_Saved_kWh': [12, 8, 15, 9, 11],
    'Water_Saved_Liters': [45, 30, 60, 50, 42],
    'Eco_Score': [78, 65, 85, 70, 80]
}
df = pd.DataFrame(data)
df
```

## 3. AI Insight Generator

```python
insights = [
    "Reduce daily water usage by 10%",
    "Use energy-efficient appliances",
    "Recycle at least 3 types of waste every week",
    "Plant a tree every month",
    "Opt for public transport to reduce carbon footprint"
]

df['AI_Insight'] = [random.choice(insights) for _ in range(len(df))]
df
```

## 4. Visualize User Sustainability Data

```python
plt.figure(figsize=(10,5))
plt.bar(df['User'], df['Energy_Saved_kWh'], color='green')
plt.title('Energy Saved by Users (kWh)')
plt.xlabel('User')
plt.ylabel('Energy Saved (kWh)')
plt.show()

plt.figure(figsize=(10,5))
plt.bar(df['User'], df['Water_Saved_Liters'], color='blue')
plt.title('Water Saved by Users (Liters)')
plt.xlabel('User')
plt.ylabel('Water Saved (Liters)')
plt.show()

plt.figure(figsize=(10,5))
plt.bar(df['User'], df['Eco_Score'], color='orange')
plt.title('Eco Score by Users')
plt.xlabel('User')
plt.ylabel('Eco Score')
plt.show()
```
